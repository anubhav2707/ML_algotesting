```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./BIOCON.csv", parse_dates=[0],index_col=0)

df
```

```python
df.columns
```

```python
df['Spread High-Open']=df['High Price']-df['Open Price']
df['Spread Open-Low']=df['Open Price']-df['Low Price']
df['Open-prev_Close']=0
for i in range(len(df)):
    df.iloc[i,df.columns.get_loc("Open-prev_Close")]=df.iloc[i][0]-df.iloc[i-1][3]
    

df
```

```python
high_low=[]
open_prev_close=[]
high_open=[]
open_low=[]
l=[25,50,100,150,0,-0.05,-25,-50]

def abc(col,name):
    for pos in range(8):
        v=0
        for i in range(len(df)):
            if l[pos]>=0:
                if df.iloc[i][col]>l[pos]:
                    v=v+1
            else:
                if df.iloc[i][col]<l[pos]:
                    v=v+1
        name.append([l[pos],int(v/len(df)*100)])
abc(10,high_low)
abc(14,open_prev_close)
abc(12,high_open)
abc(13,open_low)
```
